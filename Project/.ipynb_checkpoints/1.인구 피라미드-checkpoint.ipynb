# 인구 피라미드

# 남자 데이터 정의

```python
import pandas as pd
file_name = '201202_201202_연령별인구현황_월간.xlsx''
df_m = pd.read_excel(file_name, skiprows=3, usecols='B,E:Y')
df_m.head(3)
```

```python
df_m.drop(columns=['행정기관'], inplace=True)
df_m.iloc[0] = df_m.iloc[0].str.replace(',','', regex=True).astype(int)
```

```python
df_m.head(3)
```

# 여자 데이터 정의

```python
df_w = pd.read_excel(file_name, skiprows=3, usecols='B,AB:AV')
df_w.head(3)
```

```python
df_w.columns = df_m.columns
df_w.columns
```

```python
df_w.drop(columns=['행정기관'], inplace=True)
df_w.iloc[0] = df_w.iloc[0].str.replace(',','', regex=True).astype(int)
```

```python
df_w
```

# 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Apple SD Gothic Neo'
matplotlib.rcParams['font.size'] = 15 #글자 크기
```

```python
plt.figure(figsize=(20, 25), dpi=80)
plt.barh(df_m.columns, -df_m.iloc[0]//1000)
plt.barh(df_w.columns, df_w.iloc[0]//1000)
plt.title('2012 인구피라미드')
plt.savefig('2012_인구피라미드.png', dpi=100)
plt.show()
```
