# 데이터 선택(조건)

```python
import pandas as pd
df = pd.read_excel('score.xlsx')
df
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt]
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185, '수학'] #키 185이상들의 수학 데이터 
```

```python
df.loc[df['키'] >= 185,['이름', '수학', '과학']]
```

## 다양한 조건
### & 그리고

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')]
#키가 185이상이면서 학교가 북산고인 데이터 가져오기
```

### | 또는

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)]
#키가 170보다 작거나, 200보다 큰 데이터(row) 가져오기
```

### str 함수
문자열로 검색

```python
filt = df['이름'].str.startswith('송')
df[filt]
```

```python
filt = df['이름'].str.contains('태')
df[filt]
```

```python
df[~filt]
```

```python
langs = ["Python", "Java"]
filt = df['SW특기'].isin(langs) #SW특기가 python이거나 Java인 사람
df[filt]
```

```python
langs = ["python", "java"]
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
filt = df['SW특기'].str.contain('Java')
df[filt]
```

```python
df['SW특기'].str.contains('Java')
```

```python
df['SW특기'].str.contains('Java', na=True)
```

```python
df['SW특기'].str.contains('Java', na=False)
```

```python
filt = df['SW특기'].str.contains('Java', na=False)
df[filt]
```
