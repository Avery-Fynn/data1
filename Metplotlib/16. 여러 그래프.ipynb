## 16. 여러 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Apple SD Gothic Neo'
matplotlib.rcParams['font.size'] = 15 #글자 크기
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
#2x2에 해당하는 플랏들을 생성
fig, axs = plt.subplots(2, 2, figsize=(16,13))
fig.suptitle('graphs')

axs[0,0].bar(df['지원번호'], df['국어'], label='국어 점수', width=0.5)
axs[0,0].set_title('첫 번째 그래프')
axs[0,0].legend()
axs[0,0].set(xlabel='지원번호', ylabel='점수')
axs[0,0].set_facecolor('lightblue')
axs[0,0].grid(ls='--', lw=0.5)

axs[0,1].plot(df['지원번호'], df['수학'], label='국어 점수')
axs[0,1].plot(df['지원번호'], df['영어'], label='영어 점수')
axs[0,1].set_title('두 번째 그래프')
axs[0,1].legend()
axs[0,1].set(xlabel='지원번호', ylabel='점수')
axs[0,1].grid(ls='--', lw=0.5)

axs[1,0].barh(df['지원번호'], df['키'], color = 'indianred')
axs[1,0].set_facecolor('lightyellow')
axs[1,0].set_title('세 번째 그래프')
axs[1,0].set(xlabel='지원번호', ylabel='점수')
axs[1,0].grid(ls='--', lw=0.5)

axs[1,1].plot(df['지원번호'], df['사회'], color='green', label='사회 점수', alpha=0.6)
axs[1,1].set_title('네 번째 그래프')
axs[1,1].legend()
axs[1,1].set(xlabel='지원번호', ylabel='점수')
axs[1,1].grid(ls='--', lw=0.5)

plt.show()
```
