# 4. 스타일 

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Apple SD Gothic Neo'
matplotlib.rcParams['font.size'] = 15 #글자 크기
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=5)
```

## marker

```python
plt.plot(x,y, marker='o')
```

```python
plt.plot(x,y, marker='o', linestyle='None')
```

```python
plt.plot(x,y, marker='v', linestyle='None')
```

```python
plt.plot(x,y, marker='v', markersize=10, linestyle='None')
```

```python
plt.plot(x,y, marker='X', markersize=10, linestyle='None')
```

```python
plt.plot(x,y, marker='X', markersize=10, markeredgecolor='red', linestyle='None')
```

```python
plt.plot(x,y, marker='o', markersize=20, markeredgecolor='red', linestyle='None')
```

```python
plt.plot(x,y, marker='o', markersize=10, markeredgecolor='red', markerfacecolor='red', linestyle='None')
```

```python
plt.plot(x,y, marker='o', markersize=10, markeredgecolor='red', markerfacecolor='yellow', linestyle='None')
```

## 선 스타일

```python
plt.plot(x,y, linestyle=':')
```

```python
plt.plot(x,y, linestyle='--')
```

```python
plt.plot(x,y, linestyle='-.')
```

## 색깔

```python
plt.plot(x,y, color='pink')
```

```python
plt.plot(x,y, color='b')
```

```python
plt.plot(x,y, color='r')
```

```python
plt.plot(x,y, color='g')
```

```python
plt.plot(x,y, color='teal')
```

```python
plt.plot(x,y, color='indianred')
```

## 포맷

```python
# color, marker, linestyle
plt.plot(x,y, 'ro--')
```

```python
plt.plot(x,y, 'bv:')
```

```python
plt.plot(x,y, 'go')
```

## 축약어


```python
plt.plot(x,y, marker='o',mfc='red', ms=15, mec='blue', ls=':')
```

```python
plt.plot(x,y, marker='o',mfc='indianred', ms=10, alpha=0.5)
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5), dpi=60)
plt.plot(x,y)
```

```python
plt.figure(facecolor='#a1c3ff', figsize=(10,5), dpi=60)
plt.plot(x,y)
```
