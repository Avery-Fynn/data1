# 원 그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Apple SD Gothic Neo'
matplotlib.rcParams['font.size'] = 15 #글자 크기
```

```python
values = [30, 25, 20, 13, 10, 2] #다 더해서 100일 필요는 없음
labels = ['python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#ADD8E6', '#BC8F8F']
explode = [0.05] * 6

plt.figure(figsize=(7, 7), dpi=80)
plt.pie(values, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False,  colors=colors, explode=explode)
plt.show()
```

```python
wedgeprops = {'width':0.5, 'edgecolor':'w', 'linewidth'}
plt.figure(figsize=(7, 7), dpi=80)
plt.pie(values, wedgeprops = wedgeprops, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False,  colors=colors)
plt.show()
```

```python
def custom_autopct(pct):
#     return ('%.1f%%' % pct) if pct >= 10 else ''
    return '{:.0f}%'.format(pct) if pct >= 10 else ''

plt.figure(figsize=(7, 7), dpi=80)
plt.pie(values, pctdistance=0.75, wedgeprops = wedgeprops, labels = labels, autopct=custom_autopct, startangle=90, counterclock=False,  colors=colors)
plt.show()
```
