# 9. 막대 그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Apple SD Gothic Neo'
matplotlib.rcParams['font.size'] = 15 #글자 크기
```

```python
labels = ['Avery', 'Basta', 'Christina'] #이름
values = [190, 187, 184] #키
colors = ['r', 'g', 'b'] #색
ticks = ['1번 학생','2번 학생','3번 학생']

plt.barh(labels, values, color=colors, alpha=0.5)
plt.xlim(175, 195)
plt.show()
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('//')
bar[1].set_hatch('*')
bar[2].set_hatch('xx')
```

```python
bar = plt.bar(labels, values, width=0.3)
bar[0].set_hatch('//')
bar[1].set_hatch('*')
bar[2].set_hatch('xx')
plt.ylim(175, 195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5 , values[idx], ha='center', color='blue')
```
