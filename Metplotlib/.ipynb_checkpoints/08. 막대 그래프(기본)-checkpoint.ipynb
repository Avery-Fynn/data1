# 8. 막대 그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Apple SD Gothic Neo'
matplotlib.rcParams['font.size'] = 15 #글자 크기
```

```python
labels = ['A', 'B', 'C']
values = [190, 187, 184]

plt.bar(labels, values)
```

```python
labels = ['A', 'B', 'C']
values = [190, 187, 184]

plt.bar(labels, values, color='indianred')
```

```python
labels = ['A', 'B', 'C']
values = [190, 187, 184]
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
labels = ['A', 'B', 'C'] #이름
values = [190, 187, 184] #키
colors = ['r', 'g', 'b'] #색

plt.bar(labels, values, color=colors, alpha=0.5)
plt.ylim(180, 195)
```

```python
plt.bar(labels, values, color=colors, alpha=0.5, width=0.3)
plt.ylim(180, 195)
```

```python
labels = ['Avery', 'Basta', 'Christina'] #이름
values = [190, 187, 184] #키
colors = ['r', 'g', 'b'] #색

plt.bar(labels, values, color=colors, alpha=0.5, width=0.3)
plt.ylim(180, 195)
plt.xticks(rotation=45) #x축 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45) #y축 이름 데이터 각도를 45도로 설정
plt.show()
```

```python
labels = ['Avery', 'Basta', 'Christina'] #이름
values = [190, 187, 184] #키
colors = ['r', 'g', 'b'] #색
ticks = ['1번 학생','2번 학생','3번 학생']

plt.bar(labels, values, color=colors, alpha=0.5, width=0.3)
plt.ylim(180, 195)
plt.xticks(labels, ticks, rotation=45) #x축 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45) #y축 이름 데이터 각도를 45도로 설정
plt.show()
```
